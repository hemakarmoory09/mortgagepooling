```python
import os, json
import pandas as pd

path_to_json = '/Users/hkarmoory/Documents/Black Knight/Mortgage Pooling/Pool Optimization Data for TC v5/Pool Option Data/'
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]
#print(json_files)  # for me this prints ['foo.json']
```

```python
for file in json_files:
    with open(path_to_json+file, 'r+') as f:
        data = json.load(f)
        data['poolOptionJ'] = data.pop('Pool Option, j')
        data['poolType'] = data.pop('Pool Type')
        data['poolBalanceType'] = data.pop('Pool Balance Type')
        #print(data)
    with open(path_to_json+file, 'w') as f:
     #   filetowrite.write(data)
        json.dump(data, f, ensure_ascii=False)
       
```
